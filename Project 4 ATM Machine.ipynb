```python
import random as rd
class Atm:
    def __init__(self):
        self.pin=''
        self.__balance=rd.randrange(000,99999) # Private var it cannot be visible while pressing tab on obj.
        self.cheque=0
        self.Menu()

    
    def Menu(self):
        new=int(input("Are You New Customer Yes-->1 and No-->0:"))
        if new==1:
            print("------- Please Generate a New Pin --------")
            self.gen_pin()
        else:
            self.start()
        
    def start(self):
        choice=int(input("""
        1. Cheque Book Request
        2. Change Pin
        3. View Balance
        4. Withdraw
        5. Exit \t"""))

        if choice==1:
            self.cheque_book_req()
        elif choice==2:
            self.change_pin()
        elif choice==3:
            self.view_balance()
        elif choice==4:
            self.withdraw()
        else:
            exit()
    
    def gen_pin(self):
        new_pin=input("Enter the new Pin :")
        self.pin=new_pin
        self.start()
    
    def change_pin(self):
        pin=input("Enter Your Previous PIN:")
        if pin==self.pin:
            updated_pin=input("Enter NEW PIN:")
            print("\tYour Pin is updated Successfully")
            self.pin=updated_pin
        else:
            print("Invalid Pin TRY AGAIN!!!")
            self.change_pin()
        self.start()
    
    def view_balance(self):
        pin=input("Enter your Security Pin:")
        if pin==self.pin:
            print("Your Balance is :",self.__balance)
        else:
            print("Wrong Pin!! Retry")
        self.start()
        
    def withdraw(self):
        verify_pin=input("Enter your Pin:")
        if self.pin==verify_pin:
            amount=int(input("Enter the amount to withdraw:"))
            if amount<=self.__balance:
                self.__balance=self.__balance- amount
            else:
                print("Insufficient Balance")
        else:
            print("Wrong Pin")
        self.start()
    
    def cheque_book_req(self):
        choice=int(input("""
        Do you want to Order OR Cancel cheque book
        1--> Order
        2--> Cancel"""))
        if choice==1:
            self.cheque+=1
            print("Cheque book order Successfully")
            print(f"You have Requested for {self.cheque} Cheque book")
        else :
            self.cheque-=1
            print(f"{self.cheque} cheque book Cancel request is generated")
        self.start()

obj= Atm()
```
